## Ejercicio 2

**Caso: Una compañía desea hacer predicciones del valor anual de sus ventas totales en cierto país a partir de la relación de éstas y la renta nacional. Para investigar la relación cuenta con los siguientes datos:**

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
datos = {
    "x": np.array([189,190,208,227,239,252,257,274,293,308,316]),
    "y": np.array([402,404,412,425,429,436,440,447,458,469,469])}
```

**X representa la renta nacional en millones de euros e Y representa las ventas
de la compañía en miles de euros en el periodo que va desde 2010
hasta 2020 (ambos inclusive). Calcular:**

#### 1. La recta de regresión lineal de Y sobre X.

```python
#Ajustar modelo de regresion
x = datos["x"]
y = datos["y"]

# Etiquetar ejes
plt.scatter(x,y)
plt.xlabel("Renta Nacional Mill.Euros")
plt.ylabel("Venta compañia 2010-2020")

model = LinearRegression()
model.fit(x.reshape(-1,1),y)

#Calcular predecciones
x_pred = np.linspace(x.min(),x.max())
y_pred = model.predict(x_pred.reshape(-1,1))

# Graficar recta ajustada
plt.plot(x_pred, y_pred, color="red")

# Mostrar grafico
plt.show()
```

#### 2. El coeficiente de correlación lineal e interpretarlo.

```python
print("Intercepcion:",round(model.intercept_,4))
print("Pendiente:",round(model.coef_[0],4))
```

#### 3. Si en 2011 la renta nacional del país fue de 325 millones de euros.

**¿Cuál será la predicción para las ventas de la compañía en este año?**

```python
print("Si la renta nacional es de 325 millones de euros el resultado de la prediccion es de:",model.predict(np.array(325).reshape(-1,1))[0])
```
