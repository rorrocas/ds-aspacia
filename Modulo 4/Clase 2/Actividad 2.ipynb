```python
import numpy as np
from scipy.stats import norm
import pandas as pd
```

```python

def valor_z(start,end,step):

    z_scores = np.arange(start, end + step, step)
    # cumulative distribution function
    probabilities = norm.cdf(z_scores)
    df = pd.DataFrame({"z-score": z_scores, "probability": probabilities})
    return df
```

- p es la probabilidad de √©xito.

n=12

p(x=6) = ((12!)/(6!(12-6)!))*((0.5)^(6))*((0.5)^(12-6))

```python
factorial_12 = np.prod(np.arange(1, 12))
factorial_6 = np.prod(np.arange(1, 6))
```

```python
potencia_1 = np.power(0.5, 6)
```

```python
probabilidad = ((factorial_12)/(factorial_6*(factorial_6)))*(potencia_1)*(potencia_1)
```

```python
print(f"La probabilidad de obtener 6 aciertos es de {probabilidad:.4f}")
```

- Distribuci√≥n de probabilidad

**R: Es la funcion que describre la probabilidad de obtener un valor especifico en un conjunto de valores, al seleccionarlo de forma aleatoria.**

- Variables Aleatorias(discretas y Continuas)

**R: Las variables discretas son aquellas que dentro de una rango de valores existe un numero finito de valores mientras que las continuas toman valores infinitos en cualquier rango.**

- Distribuci√≥n Normal

**R: Es un modelo probabilistico usado normalmente para interpretar sucesos de la naturaleza que tengan una tendencia central y un variabilidad simetrica.**

- Distribuci√≥n Binomial

**R: Es un modelo para calcular las probabilidades de obtener cierta cantidad de aciertos cuando existen solo dos posibilidades como acierto y fracaso.**
