*a) Generar un DataFrame con los datos del fichero.*

```python
import pandas as pd

df_titanic = pd.read_csv("titanic.csv")
df_titanic.head(5)
```

*b) Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los nombres de sus columnas y filas, los tipos de datos de las columnas, las 10 primeras filas y las 10 últimas filas*

```python
print(f"Numero de datos: {df_titanic.shape[0]}\n")
print(f"Columnas: {df_titanic.columns.values}\n")
print(f"Tipo de datos:\n{df_titanic.dtypes}\n")
print(f"Primeras 10 filas:\n{df_titanic.head(10)}\n")
print(f"Ultimas 10 filas:\n{df_titanic.tail(10)}\n")
```

*c) Mostrar por pantalla los datos del pasajero con identificador 148.*

```python
df_titanic[df_titanic.index == 148]
```

*d) Mostrar por pantalla las filas pares del DataFrame.*

```python
df_titanic[df_titanic.index % 2 == 0]
```

*e) Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente.*

```python
sorted(df_titanic[df_titanic.Pclass == 1].Name.values)
```

*f) Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.*

```python
survived = df_titanic.Survived.value_counts("%")
print(f"Sobrevivio el {round(survived[1]*100,2)}%")
```

*g) Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.*

```python
mascara_primera_clase = df_titanic[df_titanic["Pclass"]==1]
mascara_segunda_clase = df_titanic[df_titanic["Pclass"]==2]
mascara_tercera_clase = df_titanic[df_titanic["Pclass"]==3]
```

```python
sobrevivientes_1 = mascara_primera_clase.Survived.value_counts("%")
sobrevivientes_2 = mascara_segunda_clase.Survived.value_counts("%")
sobrevivientes_3 = mascara_tercera_clase.Survived.value_counts("%")
```

```python
print(f"En primera clase sobrevivio el {round(sobrevivientes_1[1]*100,2)}%")
print(f"En segunda clase sobrevivio el {round(sobrevivientes_2[1]*100,2)}%")
print(f"En tercera clase sobrevivio el {round(sobrevivientes_3[1]*100,2)}%")
```

*h) Eliminar del DataFrame los pasajeros con edad desconocida.*

```python
df_titanic = df_titanic.dropna(subset=["Age"])
```

*i) Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.*

```python
primera_clase = df_titanic[df_titanic["Pclass"] == 1]
mujeres_primera_clase = primera_clase[primera_clase["Sex"] == "female"]
mujeres_primera_clase.Age.median()
```

*j) Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.*

```python
#df_titanic["menor"] = df_titanic.iloc[df_titanic["Age"]<18]
#df_titanic["menor"]
df_titanic_menor = df_titanic["Age"]<18
df_titanic["menor"] = df_titanic_menor
df_titanic
```

*k) Mostrar por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada clase.*

```python
menores = df_titanic[df_titanic["menor"] == True]
mayores = df_titanic[df_titanic["menor"] == False]
```

```python
print(f"Sobrevieron el {round(menores.Survived.value_counts('%')[0]*100,4)} de los menores de edad.")
print(f"Sobrevieron el {round(mayores.Survived.value_counts('%')[0]*100,4)} de los mayores de edad.")
```
