```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

np.random.seed(0)
```

```python
tata = np.random.randn(50)
plt.plot(tata)
```

```python
# np.random.randn?
```

```python
# Generar datos sintéticos con una relación lineal
pendiente = 2
intercepcion = 1
x = np.linspace(0, 10, 50)
y = pendiente * x + intercepcion + np.random.randn(50) * 2

# Graficar los datos
plt.scatter(x, y, label="datos")
plt.xlabel("x")
plt.ylabel("y")

# Ajustar un modelo de regresión lineal
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

# Calcular las predicciones usando el modelo ajustado
x_pred = np.linspace(0, 10, 100)
y_pred = model.predict(x_pred.reshape(-1, 1))

# Graficar la recta ajustada
plt.plot(x_pred, y_pred, color="red", label="modelo")
plt.legend()
# Mostrar la gráfica
plt.show()
```

```python
import pickle

# Guardar el modelo en un archivo
with open("regre_lineal_model.pkl", "wb") as f:

    pickle.dump(model, f)
```
