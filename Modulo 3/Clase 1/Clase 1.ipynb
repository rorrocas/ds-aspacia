### Visualizacion de datos

```python
import numpy as np
import pandas as pd
```

```python
df_titanic = pd.read_csv(r"C:\Users\Rodrigo\Curso Aspasia\Modulo 3\titanic.csv")
```

```python
df_titanic.dtypes
```

```python
frecuencia = df_titanic["Sex"].value_counts()
```

```python
import matplotlib as plt
```

```python
#Frecuencia acumulada para Pclass
frecuencia = df_titanic.Pclass.value_counts()
frecuencia_norma = df_titanic.Pclass.value_counts(normalize=True)
```

```python
frecuencia_acumulada = frecuencia.cumsum()
frecuencia_acumulada_norma = frecuencia_norma.cumsum()
print(frecuencia_acumulada)
print(frecuencia_acumulada_norma)
```

# Variables numericas

```python
# MEdia, mediana y moda
print(df_titanic.Age.mean())
print(df_titanic.Age.median())
print(df_titanic.Age.mode())
```

```python
# Calcular desviacion estandar, varianza, rango y coeficiente de variacion
print(df_titanic.Age.std())
print(df_titanic.Age.var())
print(df_titanic.Age.max() - df_titanic.Age.min())
print(df_titanic.Age.std() / df_titanic.Age.mean())
```

```python
# Plotear histograma de age
df_titanic.Age.plot(kind='hist', bins=20)
```

```python
# Box plot de Age
df_titanic.Age.plot(kind='box')
```

```python
#Calcular quantiles de Age
df_titanic.Age.quantile(q=[0, 0.25, 0.5, 0.75, 1])
```

```python
fig, axs = plt.pyplot.subplots(
    ncols=2, nrows=2,
    figsize=(8,8)
)
axs[0,0].set_xlabel("Edad")
axs[0,1].set_xlabel("Fare")

axs[1,0].set_xlabel("Fare")
axs[1,1].set_xlabel("Fare")

df_titanic.Age.plot(kind="box", ax=axs[0,0])
df_titanic.Age.plot(kind="hist", ax=axs[0,1])

df_titanic.Fare.plot(kind="box", ax=axs[1,0])
df_titanic.Fare.plot(kind="hist", ax=axs[1,1])
```
