```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

np.random.seed(0)
```

```python
pendiente = 2
interseccion = 1
x = np.linspace(00,10,50)
y = pendiente * x  + interseccion + np.random.randn(50) * 2
```

```python
# Graficar datos
plt.plot(y)
```

```python
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")

#Ajustar modelo de regresion
model = LinearRegression()
model.fit(x.reshape(-1,1),y)

#Calcular predecciones
x_pred = np.linspace(0,10,100)
y_pred = model.predict(x_pred.reshape(-1,1))

# Graficar recta ajustada
plt.plot(x_pred, y_pred, color="red")

# Mostrar grafico
plt.show()
```

```python
print("Intercepcion:",model.intercept_)
print("Pendiente:",model.coef_[0])
```

```python
from sklearn.metrics import r2_score
```

```python
# Calcular el coeficiente de determinacin R^2
r2 = r2_score(y, model.predict(x.reshape(-1,1)))
print("Coeficiente de determinacin R^2:", r2)
```
