# Ejercicio 1

**En una clínica infantil se ha ido anotando, durante un mes, el número de
metros que cada niño anda, seguido y sin caerse, el primer día que comienza
a caminar, obteniéndose la tabla de información adjunta:**

```python
# Importar librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Crear df
df = pd.DataFrame({
    "metros":[1,2,3,4,5,6,7,8],
    "niños": [2,6,10,5,10,3,2,2]
})
df
```

### A) Tabla de Frecuencia. Diagrama de Barras para las frecuencias absolutas, relativas y acumuladas.

```python
# Creando df Transformando tabla de  frecuencia
etiquetas = df['metros']
frecuencias = df['niños']
valores = []
for i in range(len(etiquetas)):
    for j in range(frecuencias[i]):
        valores.append(int(etiquetas[i]))
df_valores = pd.DataFrame(valores, columns=['Metros'])
df_valores.index.name= "index"
```

```python
frecuencia = df_valores.Metros.value_counts()
```

```python
# Creando grafico de frecuencia
plt.bar(df['metros'], frecuencias)

plt.xlabel('Metros')
plt.ylabel('Cant. Niños')
plt.title("Frecuencia Absoluta")

# Mostrar el gráfico
plt.show()
```

```python
# frecuencia relativa
frec_relativas = frecuencias / frecuencias.sum()
# frecuencia acumulada
frec_acumuladas = np.cumsum(frec_relativas)

# Creando grafico de frecuencia acumulada
plt.plot(frec_acumuladas.index, frec_acumuladas)
plt.title("Frecuencia acumulada")
plt.xlabel("Valores")
plt.ylabel("Frecuencia acumulada")
```

```python
# Crear un gráfico de barras para las frecuencias relativas
plt.bar(frec_relativas.index, frec_relativas)
plt.title("Frecuencia relativa")
plt.xlabel("Valores")
plt.ylabel("Frecuencia relativa")
```

## B) Mediana, media aritmética, moda y cuartiles.

```python
mediana = df_valores.median()
media = df_valores.mean()
moda = df_valores.mode()
cuartiles = df_valores.quantile(q=[0, 0.25, 0.5, 0.75, 1])
print("Mediana: ",mediana[0], "Media: ", media[0], f"Moda: Valores {moda.Metros[0]} y {moda.Metros[1]}", "\n", "Cuartiles:\n", cuartiles)
```

## C) Varianza y desviación Estándar.

```python
varianza = df_valores.var()
desviacion_estandar = df_valores.std()
print("Varianza: ", round(varianza[0],4), "Desviacion Estandar: ", round(desviacion_estandar[0],4))
```

## D) ¿Entre qué dos valores se encuentra, como mínimo, el 75% de las observaciones?

```python
print(df_valores["Metros"].quantile(0), df_valores["Metros"].quantile(0.75))
```

**R: Entre 1 y 5 se encuentra el 75% de los datos**
