#### 1. Construya una tabla fijando “continent”, con las funciones vistas en clase

```python
tabla = gapminder.pivot_table(
    values='pop',
    columns='continent',
    aggfunc='mean'
)
tabla
```

```python
tabla = gapminder.pivot_table(
    values='pop',
    columns='continent',
    aggfunc='max'
)
tabla
```

```python
tabla = gapminder.pivot_table(
    values='pop',
    columns='continent',
    aggfunc='min'
)
tabla
```

```python
tabla = gapminder.pivot_table(
    values='pop',
    columns='continent',
    aggfunc='median'
)
tabla
```
