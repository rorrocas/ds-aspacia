# Ejercicio 6

- ¿Que es Git?

    **R: Es un programa que sirve para controlar las versiones en el desarrollo de un programa/codigo**

- ¿Para que se puede usar?

    **R:Para guardar versiones de nuestro codigo, para trabajar en equipos de trabajo o para crear versiones paralelas al codigo.**
    
- ¿Que se hace con los entornos en este espacio?

    **R:Los entornos de trabajo pueden conectarse a un mismo git y compartir asi el progreso de cada trabajador, creadno bifurcaciones apra evitar conflicto en el codigo y combinandolas cuando este funciones y se quiera poner en desarrollo**
