## Probability Distributions

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the possible outcomes of rolling a die
outcomes = [1, 2, 3, 4, 5, 6]

# Define the probabilities of each outcome
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

# Create a probability mass function
pmf = np.array(probabilities)

# Plot the PMF
plt.stem(outcomes, pmf)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function of Rolling a Die')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of possible values of a continuous random variable
x = np.linspace(-5, 5, 1000)

# Define the probability density function
pdf = (1/np.sqrt(2*np.pi))*np.exp(-x**2/2)

# Plot the PDF
plt.plot(x, pdf)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of a Standard Normal Random Variable')
plt.show()
```

```python
import numpy as np

# Define the possible outcomes of rolling a die
outcomes = [1, 2, 3, 4, 5, 6]

# Define the probabilities of each outcome
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

# Calculate the mean of the random variable
mean = np.sum(np.array(outcomes)*np.array(probabilities))

# Calculate the variance of the random variable
variance = np.sum((np.array(outcomes)-mean)**2*np.array(probabilities))

print('Mean:', mean)
print('Variance:', variance)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the possible outcomes of rolling a die
outcomes = [1, 2, 3, 4, 5, 6]

# Define the probabilities of each outcome
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

# Calculate the cumulative distribution function
cdf = np.cumsum(probabilities)

# Plot the CDF
plt.step(outcomes, cdf)
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of Rolling a Die')
plt.show()
```
