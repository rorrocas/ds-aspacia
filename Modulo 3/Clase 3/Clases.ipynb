```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Diferentes plots
- Diagramas de dispersión o puntos.
- Diagramas de barras.
- Histograma.
- Diagramas de caja y bigotes.

#### Diagrama de Dispersión

```python
# Generar datos aleatorios para x e y
x = np.random.randn(100)
y = np.random.randn(100)

# Crear un diagrama de dispersión
fig, ax = plt.subplots()
ax.scatter(x, y)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Diagrama de Dispersión')

# Mostrar el gráfico
plt.show()
```

#### Diagrama de barras

```python
# Generar datos aleatorios para el eje x e y
x = np.array(['A', 'B', 'C', 'D', 'E'])
y = np.random.rand(5)

# Crear un diagrama de barras
fig, ax = plt.subplots()
ax.bar(x, y)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Diagrama de Barras')

# Mostrar el gráfico
plt.show()
```

#### Histograma

```python
# Generar datos aleatorios para el eje x
x = np.random.randn(1000)

# Crear un histograma
fig, ax = plt.subplots()
ax.hist(x, bins=20)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma')

# Mostrar el gráfico
plt.show()
```

#### Diagrama de caja

```python
# Generar datos aleatorios para el eje x
x = np.random.randn(1000)

# Crear un diagrama de caja
fig, ax = plt.subplots()
ax.boxplot(x)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_title('Diagrama de Caja y Bigotes')

# Mostrar el gráfico
plt.show()
```

```python
df = pd.DataFrame(
    {'x': [1, 2, 3, 4, 5, 6, 7, 8],
     'w': [2, 6, 10, 5, 10, 3, 2, 2]}
)

weighted_median = np.average(df['x'], weights=df['w'])
print(weighted_median)
```

```python
weighted_avg = (df['x'] * df['w']).sum() / df['w'].sum()

print(weighted_avg)
```

```python
std_w = np.sqrt(
    np.average(
        (df['x'] - weighted_avg)**2,
        weights=df['w']
    )
)
print('Weighted Standard Deviation:', std_w)
```
