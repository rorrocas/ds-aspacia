```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

titanic.sample()
```

### 1. Confeccione un gr√°fico de barras para visualizar hombres y mujeres que han sobrevivido.

```python
ax = sns.countplot(
    x="sex",
    data=titanic,
    hue="survived",
    palette=["red","blue"]
)
plt.title("Sobrevivencia por sexo")
plt.xlabel('Genero')
plt.ylabel('Numero de personas')
```

### 3.- Graficos agrupados

```python
titanic_sobrevivientes = titanic[titanic["survived"]==1]
```

```python
#POR CLASE
ax = sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    hue="pclass",
    palette="colorblind"
)
```

```python
# Por embarque
ax = sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    hue="embarked",
    palette="colorblind"
)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,5))

sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    hue="pclass",
    palette="colorblind", ax = axs[0]
)

sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    hue="embarked",
    palette="colorblind", ax=axs[1]
)
```
