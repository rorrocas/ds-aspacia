```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

import numpy as np
```

## Diferentes plots
* Diagramas de dispersión o puntos.
* Diagramas de barras.
* Histograma.
* Diagramas de caja y bigotes.

#### Diagrama de Dispersión

```python
# Generar datos aleatorios para x e y
x = np.random.randn(100)
y = np.random.randn(100)

# Crear un diagrama de dispersión
fig, ax = plt.subplots(
    nrows=1, ncols=1,
    figsize=None
)

ax.scatter(x, y)

# # Establecer las etiquetas de los ejes y el título
ax.set_xlabel('Variable X')
ax.set_ylabel('Variable Y')
ax.set_title('Diagrama de Dispersión')

# Mostrar el gráfico
plt.show()
```

### Ejemplo de varios plots en una misma figura

```python
# Crear 4 variables para hacer 4 plots de dispersion
x2 = np. linspace(0, 10*np.pi, 100)
y2 = np.sin(x2)

x1 = np.random.randn(100)
y1 = np.random.randn(100)

fig, axs = plt.subplots(
    nrows=2, ncols=2,
    # figsize=(10, 10),
)

axs[0, 0].scatter(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].scatter(x1, y2)
axs[1, 1].hist(y1)

# set tight layout with fig
fig.tight_layout()
fig.savefig('figura.pdf')

plt.show()
```

#### Diagrama de Barras

```python
# Generar datos aleatorios para el eje x e y
x = np.array(['A', 'B', 'C', 'D', 'E'])
y = np.random.rand(5)

# Crear un diagrama de barras
fig, ax = plt.subplots()
ax.bar(x, y)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Diagrama de Barras')

# Mostrar el gráfico
plt.show()
```

#### Histograma

```python
# Generar datos aleatorios para el eje x
x = np.random.randn(1000)

# Crear un histograma
fig, ax = plt.subplots()
ax.hist(x, bins=50)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma')

# Mostrar el gráfico
plt.show()
```

#### Diagrama de Caja

```python
# Generar datos aleatorios para el eje x
x = np.random.randn(5000)

# Crear un diagrama de caja
fig, ax = plt.subplots()
ax.boxplot(x)

# Establecer las etiquetas de los ejes y el título
ax.set_xlabel('X')
ax.set_title('Diagrama de Caja y Bigotes')

# Mostrar el gráfico
plt.show()
```

### Estilo de los plots

```python
# Generar datos sintéticos
x = np.linspace(0, 2*np.pi, 100)
y1 = np.cos(2*x)
y2 = np.exp(-x)
y3 = np.sin(x) * np.exp(-x)

# Crear el objeto de figura y ejes
fig, ax = plt.subplots(
    figsize=None
)

# Graficar los datos con diferentes colores y marcadores
ax.plot(
    x, y1,
    color='red', marker='o', linestyle='-',
    label='Coseno'
)

ax.plot(
    x, y3,
    color='blue', marker='s', linestyle='--',
    label='Onda amortiguada'
)

# Añadir etiquetas de los ejes y título
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Oscilaciones amortiguadas')

# Añadir una leyenda
ax.legend(loc='upper right')

# Establecer límites en los ejes x e y
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.2, 1.2)


# Añadir rejilla
ax.grid()

# Personalizar los ticks de los ejes
ax.set_xticks(np.arange(0, 2*np.pi, 1))
ax.minorticks_on()
ax.set_yticks(np.arange(-1, 1.1, 0.5))

# Añadir una anotación al máximo
ax.annotate(
    'Máximo',
    # xy=(xmax, ymax),
    xy=(np.pi, 1),
    xytext=(0.5, 0.8),
    # textcoords='offset points',
    arrowprops=dict(
        facecolor='black',
        mutation_scale=200
    )
)

# Añadir etiquetas de los ejes y título

# Mostrar la gráfica
plt.show()
```
